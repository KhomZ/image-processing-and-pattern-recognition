## Algorithm: High Pass Filter(HPF)


```
1. Read an image which is our source.
2. Define a High Pass Filter. (Eg. sharpening filter with size k*l)
3. Apply Convolution between source image and kernel using cv2.filter2D() function.

```
Source code for HPF is as below:

````python
# HPF
import cv2
import numpy as np
import matplotlib.pyplot as plt
````

````python
# read image
img2 = cv2.imread('sophia.jpg')
# m,n = img2.shape
# print(m,n)
plt.imshow(img2)
# plt.set_cmap('gray')
````

````python
# edge detection filter
kernel = np.array([[0.0, -1.0, 0.0],
                   [-1.0, 4.0, -1.0],
                   [0.0, -1.0, 0.0]])

kernel = kernel / (np.sum(kernel) if np.sum(kernel) != 0 else 1)

# filter the source image
img2_result = cv2.filter2D(img2,-1,kernel)

plt.imshow(img2_result)

# save the result image
cv2.imwrite('result.jpg', img2_result)
````
